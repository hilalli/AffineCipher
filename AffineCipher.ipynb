**Affine Cipher Encryption Part**

```python
def encrypt_affine(plain_text, a, b):
    encrypted_text = ""
    m = 26  # Assuming the alphabet size is 26 (English alphabet)

    for char in plain_text:
        if char.isalpha():
            is_upper = char.isupper()
            char_index = ord(char) - ord('A' if is_upper else 'a')

            # Apply the affine transformation
            encrypted_char_index = (a * char_index + b) % m

            # Convert back to a letter
            encrypted_char = chr(encrypted_char_index + ord('A' if is_upper else 'a'))

            encrypted_text += encrypted_char
        else:
            encrypted_text += char

    return encrypted_text

# Example Usage:
plaintext = "HELLO"
a = 5
b = 10
cipher_text = encrypt_affine(plaintext, a, b)
print(f"Encrypted Text: {cipher_text}")
```

**Affine Cipher Decryption Part**

```python
def mod_inverse(a, m):
    # Calculate the modular multiplicative inverse of a mod m
    for i in range(1, m):
        if (a * i) % m == 1:
            return i
    return None

def decrypt_affine(cipher_text, a, b):
    decrypted_text = ""
    m = 26  # Assuming the alphabet size is 26 (English alphabet)

    a_inverse = mod_inverse(a, m)
    if a_inverse is None:
        raise ValueError("Invalid key: 'a' must be chosen such that (a * a^-1) mod m = 1")

    for char in cipher_text:
        if char.isalpha():
            is_upper = char.isupper()
            char_index = ord(char) - ord('A' if is_upper else 'a')

            # Apply the inverse of the affine transformation
            decrypted_char_index = (a_inverse * (char_index - b)) % m

            # Convert back to a letter
            decrypted_char = chr(decrypted_char_index + ord('A' if is_upper else 'a'))

            decrypted_text += decrypted_char
        else:
            decrypted_text += char

    return decrypted_text

# Example Usage:
decrypted_text = decrypt_affine(cipher_text, a, b)
print(f"Decrypted Text: {decrypted_text}")
```
